In Netflix, There are more No of Movies than TV shows

```python
# Top 10 countries with the highest no of views
df5['country'].value_counts().head(10)
```

```python
#Checking the no of contents watched by the top 10 countries
plt.figure(figsize = (12, 6))
sns.countplot(y='country', order = df5['country'].value_counts().index[0:10], data = df5)
plt.title("Country wise content on Netflix")
```

```python
movie_countries = df5[df5['type'] == 'Movie']
tv_show_countries = df5[df5['type'] == 'TV Show']
```

```python
plt.figure(figsize = (12, 6))
sns.countplot(y='country', order = df5['country'].value_counts().index[0:10], data = movie_countries)
plt.title("Top 10 countries Movies shown on Netflix")


plt.figure(figsize = (12, 6))
sns.countplot(y='country', order = df5['country'].value_counts().index[0:10], data = tv_show_countries)
plt.title("Top 10 countries TV shows shown on Netflix")
```
